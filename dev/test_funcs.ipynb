```python
import matplotlib.pyplot as plt
```

```python
def plot_cloud(wordcloud):
    # Set figure size
    plt.figure(figsize=(40, 20))
    # Display image
    plt.imshow(wordcloud) 
    # No axis details
    plt.axis("off")
    plt.tight_layout(pad=0);
```
